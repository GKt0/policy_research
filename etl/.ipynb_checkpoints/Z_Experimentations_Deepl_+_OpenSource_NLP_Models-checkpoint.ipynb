1) Detecting source language was possible but we cancelled the idea because it misses to much on single word inputs.
2) Manage source selected language to target selected language(s)

```python
AUTH_KEY = "your_key"
```

```python
#User's inputs
selected_source_lan = "English"
query = "tax evasion"
selected_target_lan = ["Dutch", "Polish", "French"]
```

```python
#Retreive languages CODEs in order to query
source_lang = source_langs[selected_source_lan]
print(source_lang)

if(len(selected_target_lan)>=1):
    target_lan = []
    for i in range(0,len(selected_target_lan)):
        target_lan.append(target_langs[selected_target_lan[i]])
else:
   target_lan = target_langs[selected_target_lan[i]] 
print(target_lan)  
```

3) Translate

```python
#Few usage: french translation : cleaning articles: le, la, les, l', un, une, des
def cleaning(str_in):
    #regex
    result = re.sub(r"\b(?:l['e]|la|un[e]?|le)\s+", "", str_in, flags=re.IGNORECASE)
    return result

translator = deepl.Translator(AUTH_KEY)
translation = []

#include single or multiple translations
for code_lan in target_lan:
    translation.append(translator.translate_text(query, target_lang=code_lan, source_lang=source_lang))

for t in translation:
    print(t.text)

#print(translation.detected_source_lang)
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)
```

We can now open & load these indexed texts and perform a search

```python
embeddings.load("9k9_data_text_indexed")
res = embeddings.search("échelonnement agent", 10)


for item_res in res:
    print(f"Text: {txtai_data[item_res[0]][1]}")
    print(f"Similarity: {item_res[1]} \n")
    print("====================")
```
